### Defining the ticker list for financial statements group.

```python
tickers = ['aapl', 'msft', 'goog', 'aig', 'pcg', 'mcd', 'bcs', 'nvda', 'intc', 'amzn']
```

excluding pypl as for this stock only 2 financial statements were available(were in the right range chronologically), which is too short for the forecast.

```python
stock_dfs = {}
for t in tickers:
    try:
        # creating separate dataframes for each stock
        stock_dfs[t]= df.loc[(df['Company '] == t.upper())]
        # saving to csv
        stock_dfs[t].to_csv(f'../data/clean_csv/fs_{t}.csv')
        stock_dfs[t].info()
    except:
        print(f'{t} does not exist in this list')
    
```

```python
stock_p_dfs = {}
for t in tickers:
    stock_p_dfs[t] = pd.read_csv(f'../data/clean_csv/{t}.csv')
    stock_p_dfs[t]["Date"] = pd.to_datetime(stock_p_dfs[t]["Date"])
    stock_p_dfs[t] = pd.merge(
        left=stock_p_dfs[t], 
        right=stock_dfs[t],
        how='left',
        left_on='Date',
        right_on='Date',
    )
    stock_p_dfs[t].drop(columns = ['Year', 'Company ','Category', 'Company','Unnamed: 0'],inplace=True)
    stock_p_dfs[t].ffill(inplace=True)
    stock_p_dfs[t].dropna(inplace=True, ignore_index=True)
    stock_p_dfs[t].to_csv(f'../data/clean_csv/{t}_p_fs.csv')
    print(stock_p_dfs[t].head(5))
```
